```python
def attributes_to_json(data) -> str:

  data_ret = {}

  for attributes in data:
    try:
      data_parsed = attributes.split(":")
      data_ret[data_parsed[0]] = data_parsed[1]
    except:
      continue

  return str(data_ret)
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

def clean_and_extract_words(sentence):
  sentence = sentence.lower()
  stop_words = set(stopwords.words('english'))
  word_tokens = word_tokenize(sentence)
  filtered_sentence = [w for w in word_tokens if not w in stop_words and w.isalnum()]


  return filtered_sentence
```

```python
def get_model(input_shape_feat = X_tfidf_feat.shape[1], input_shape_search = X_tfidf_search.shape[1]):

  input_feat = tf.keras.layers.Input(shape=(input_shape_feat,), dtype='float32', name='feature_input')
  input_search = tf.keras.layers.Input(shape=(input_shape_search,), dtype='float32', name='search_input')

  input_dense_feat = tf.keras.layers.Dense(256, activation='relu')(input_feat)
  batch_norm_feat = tf.keras.layers.BatchNormalization()(input_dense_feat)
  dropout_feat = tf.keras.layers.Dropout(0.2)(batch_norm_feat)
  input_dense_search = tf.keras.layers.Dense(128, activation='relu')(input_search)
  batch_norm_search = tf.keras.layers.BatchNormalization()(input_dense_search)
  dropout_search = tf.keras.layers.Dropout(0.2)(batch_norm_search)

  concat_layer = tf.keras.layers.Concatenate()([dropout_feat, dropout_search])

  dense_layer_1 = tf.keras.layers.Dense(128, activation='relu')(concat_layer)
  batch_norm_1 = tf.keras.layers.BatchNormalization()(dense_layer_1)
  dropout_1 = tf.keras.layers.Dropout(0.2)(batch_norm_1)
  dense_layer_2 = tf.keras.layers.Dense(64, activation='relu')(dropout_1)
  batch_norm_2 = tf.keras.layers.BatchNormalization()(dense_layer_2)
  dropout_2 = tf.keras.layers.Dropout(0.2)(batch_norm_2)
  dense_layer_3 = tf.keras.layers.Dense(32, activation='relu')(dropout_2)
  batch_norm_3 = tf.keras.layers.BatchNormalization()(dense_layer_3)

  output_layer = tf.keras.layers.Dense(1, activation='linear')(batch_norm_3)

  model = tf.keras.models.Model(inputs=[input_feat, input_search], outputs=output_layer)
  return model
```

```python
def get_model(input_shape_feat = 512, input_shape_search = 512):

  input_feat = tf.keras.layers.Input(shape=(input_shape_feat,), dtype='float32', name='feature_input')
  input_search = tf.keras.layers.Input(shape=(input_shape_search,), dtype='float32', name='search_input')

  input_dense_feat = tf.keras.layers.Dense(256, activation='relu')(input_feat)

  input_dense_search = tf.keras.layers.Dense(256, activation='relu')(input_search)

  concat_layer = tf.keras.layers.Concatenate()([input_dense_feat, input_dense_search])

  dense_layer_1 = tf.keras.layers.Dense(256, activation='relu')(concat_layer)
  batch_norm_1 = tf.keras.layers.BatchNormalization()(dense_layer_1)
  dense_layer_2 = tf.keras.layers.Dense(128, activation='relu')(batch_norm_1)
  batch_norm_2 = tf.keras.layers.BatchNormalization()(dense_layer_2)
  dense_layer_3 = tf.keras.layers.Dense(64, activation='relu')(batch_norm_2)
  batch_norm_3 = tf.keras.layers.BatchNormalization()(dense_layer_3)

  output_layer = tf.keras.layers.Dense(1, activation='linear')(batch_norm_3)

  model = tf.keras.models.Model(inputs=[input_feat, input_search], outputs=output_layer)
  return model
```
